### Import CIFAR-10-Small and Display Some Images

```python
from sklearn.datasets import fetch_openml

images_cifar10s = fetch_openml('CIFAR_10_small')
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
for i in range(15):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    # By default, it is C-contiguous (which is in fact row-major), 
    # but for matplotlib, you'll want the red channel values in [:,:,0]. 
    # That means you should read that data in in Fortran order so that 
    # it first fills the "columns" (in this 3D context).
    # https://stackoverflow.com/questions/28005669/how-to-view-an-rgb-image-with-pylab
    # also need to swap x, y axes (tranpose on x and y only) to present images correctly
    #
    # need to cast the RGB values as np.uint8 for matplotlib to plot correctly
    # https://stackoverflow.com/questions/49643907/clipping-input-data-to-the-valid-range-for-imshow-with-rgb-data-0-1-for-floa
    plt.imshow(np.reshape(images_cifar10s['data'][i], (32,32,3), order='F').swapaxes(0,1).astype(np.uint8))
plt.show()
```

### Train-Test Split 3/4 - 1/4

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import log_loss

# train-test split 3/4 - 1/4
X = images_cifar10s['data']
y = images_cifar10s['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### Multiclass Logistic Regression CV Model w/ L1 Penalty

```python
# Multiclass Logistic Regression CV Model with L1 penalty
logreg_cv_l1 = LogisticRegressionCV(Cs=[0.001, 0.01, 0.1, 1, 10, 100], cv=5, tol=1e-2, penalty='l1', multi_class='multinomial', 
                                 solver='saga', n_jobs=-1).fit(X_train, y_train)
print('Done with training')
```

```python
# Test Data Log Loss for Multiclass Logistic Regression CV w/ L1 penalty
predicted_prob = logreg_cv_l1.predict_proba(X_test)
print('Test Data Log Loss for Multiclass Logistic Regression w/ L1 penalty')
print(log_loss(y_test, predicted_prob), end='\n')
```

```python
# Train Data Log Loss for Multiclass Logistic Regression CV w/ L1 penalty
predicted_prob = logreg_cv_l1.predict_proba(X_train)
print('Train Data Log Loss for Multiclass Logistic Regression w/ L1 penalty')
print(log_loss(y_train, predicted_prob))
```

### Multiclass Logistic Regression CV Model w/ L2 Penalty

```python
# Multiclass Logistic Regression CV Model with L2 penalty
logreg_cv_l2 = LogisticRegressionCV(Cs=[0.001, 0.01, 0.1, 1, 10, 100], cv=5, tol=1e-2, penalty='l2', multi_class='multinomial', 
                                 solver='saga', n_jobs=-1).fit(X_train, y_train)
print('Done with training')
```

```python
# Test Data Log Loss for Multiclass Logistic Regression CV w/ L2 penalty
predicted_prob = logreg_cv_l2.predict_proba(X_test)
print('Test Data Log Loss for Multiclass Logistic Regression w/ L2 penalty')
print(log_loss(y_test, predicted_prob), end='\n')
```

```python
# Train Data Log Loss for Multiclass Logistic Regression CV w/ L2 penalty
predicted_prob = logreg_cv_l2.predict_proba(X_train)
print('Train Data Log Loss for Multiclass Logistic Regression w/ L2 penalty')
print(log_loss(y_train, predicted_prob), end='\n')
```

# Problem 3: Revisiting Logistic Regression and MNIST
Here we throw the kitchen sink of classical ML (i.e.  pre-deep learning) on MNIST.
- Use Random Forests to try to get the best possible test accuracy on MNIST. Use CrossValidation to find the best settings.   How well can you do? You should use the accuracy metric to compare to logistic regression. What are the hyperparameters of   your best model?
- Use Gradient Boosting to do the same. Try your best to tune your hyper parameters. What are the hyperparameters of your     best model?

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV
from pprint import pprint

# split the data
X = images_mnist['data']
y = images_mnist['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Random Forests

```python
# create Random Forest Classifier as the model for Random Searching
rf_mnist = RandomForestClassifier()

# good hyper parameters to tune
# n_estimators = n of trees
# max_features = max number of features considered for splitting a node
# max_depth = max number of levels in each decision tree
# min_samples_split = min number of data points placed in a node before the node is split
# min_samples_leaf = min number of data points allowed in a leaf node
# bootstrap = method for sampling data points (with or without replacement)

# my hyperparameters for RandomSearchCV on Random Forests
model_params = {
    'n_estimators': [100,200,300,400,500],
    'max_depth': [20,40,60,80,100],
    'min_samples_split': [1,2,4,8],
    'min_samples_leaf': [1,2,4,8]
}

# set up random search meta-estimator
# this will train 100 models over 3 folds of cross validation (300 models total max)
rsCV_mnist = RandomizedSearchCV(rf_mnist, model_params, n_iter=100, cv=3, random_state=42, n_jobs=-1, verbose=10)

# train the random search meta-estimator to find the best model out of 100 candidates
# fit returns the best model from all the ones trained from
rsCV_mnist_model = rsCV_mnist.fit(X_train, y_train)

# print winning set of hyperparameters
pprint(rsCV_mnist_model.best_estimator_.get_params())
```

```python
from sklearn.metrics import accuracy_score

preds = rsCV_mnist_model.predict(X_test)
acc = accuracy_score(y_test, preds)
print(f"MNIST Random Forest Accuracy: {acc}")
```

### Gradient Boosting

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV

# dtrain = xgb.DMatrix(X_train, label=y_train)
# dtest = xgb.DMatrix(X_test, label=y_test)

# parameter grid for XGBoost
params = {
    'n_estimators':[100,200,300,400,500],
    'learning_rate': [0.2,0.3,0.4],
    'max_depth': [3,4,5]
}
clf = xgb.XGBClassifier(objective='multi:softmax', n_jobs=-1, verbosity=2)
xgb_rsCV_mnist = RandomizedSearchCV(clf, params, n_iter=50, n_jobs=-1, cv=3, verbose=10, random_state=42)

xgb_rsCV_mnist_model = xgb_rsCV_mnist.fit(X_train, y_train)
pprint(xgb_rsCV_mnist_model.best_estimator_.get_params())
```

```python
pprint(xgb_rsCV_mnist_model.best_estimator_.get_params())
```

```python
preds = xgb_rsCV_mnist_model.predict(X_test)
acc = accuracy_score(y_test, preds)
print(f"MNIST Gradient Boost Accuracy: {acc}")
```

# Problem 4: Revisiting Logistic Regression and CIFAR-10
As before, we’ll throw the kitchen sink of classical ML (i.e. pre-deep learning) on CIFAR-10. Keep in mind that CIFAR-10 is a few times larger.
- What is the best accuracy you can get on the test data, by tuning Random Forests? What are the hyperparameters of your best model?
- What is the best accuracy you can get on the test data, by tuning any model including Gradient boosting? What are the hyperparameters of your best model?

```python
from sklearn.datasets import fetch_openml

images_cifar10 = fetch_openml('CIFAR_10')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV
from pprint import pprint

# split the data
X = images_cifar10['data']
y = images_cifar10['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Random Forests

```python
# create Random Forest Classifier as the model for Random Searching
rf_cifar10 = RandomForestClassifier()

# my hyperparameters for RandomSearchCV on Random Forests
model_params = {
    'n_estimators': [100,200,300,400,500],
    'max_depth': [20,40,60,80,100],
    'min_samples_split': [1,2,4,8],
    'min_samples_leaf': [1,2,4,8]
}

# set up random search meta-estimator
# this will train 100 models over 3 folds of cross validation (300 models total max)
rsCV_cifar10 = RandomizedSearchCV(rf_cifar10, model_params, n_iter=100, cv=3, random_state=42, n_jobs=-1, verbose=10)

# train the random search meta-estimator to find the best model out of 100 candidates
# fit returns the best model from all the ones trained from
rsCV_cifar10_model = rsCV_cifar10.fit(X_train, y_train)

# print winning set of hyperparameters
pprint(rsCV_cifar10_model.best_estimator_.get_params())
```

```python
from sklearn.metrics import accuracy_score

preds = rsCV_cifar10_model.predict(X_test)
acc = accuracy_score(y_test, preds)
print(f"CIFAR-10 Random Forest Accuracy: {acc}")
```

### Gradient Boosting

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV


# A parameter grid for XGBoost
params = {
    'n_estimators':[20,30,40,50],
    'learning_rate':[0.2,0.3,0.4],
    'max_depth':[3,4,5]
}

clf = XGBClassifier(objective='multi:softmax')
xgb_rsCV_cifar10 = RandomizedSearchCV(clf, param_distributions=params, n_iter=50, n_jobs=4, cv=3, verbose=10, random_state=42)

xgb_rsCV_cifar10_model = xgb_rsCV_cifar10.fit(X_train, y_train)
```

```python
pprint(xgb_rsCV_cifar10_model.best_estimator_.get_params())
```

```python
from sklearn.metrics import accuracy_score

preds = xgb_rsCV_cifar10_model.predict(X_test)
acc = accuracy_score(y_test, preds)
print(f"CIFAR-10 Gradient Boost Accuracy: {acc}")
```
